#### Q10. How many sharks participated in this show and What were their names? #### 

```python
sharks=shark_tank.columns[shark_tank.columns.str.contains('present')==True].str.split('_')
```

```python
sharks_name=[]
numberOfSharks=0;
for shark in sharks:
    sharks_name.append(shark[0])
    numberOfSharks+=1
    
print(f"Name of the Sharks:{sharks_name}")
print(f"Number of Sharks:{numberOfSharks}")
```

```python
ashneer_equity=shark_tank[shark_tank["ashneer_deal"]==1]["equity_per_shark"].sum()
anupam_equity=shark_tank[shark_tank["anupam_deal"]==1]["equity_per_shark"].sum()
aman_equity=shark_tank[shark_tank["aman_deal"]==1]["equity_per_shark"].sum()
namita_equity=shark_tank[shark_tank["namita_deal"]==1]["equity_per_shark"].sum()
vineeta_equity=shark_tank[shark_tank["vineeta_deal"]==1]["equity_per_shark"].sum()
peyush_equity=shark_tank[shark_tank["peyush_deal"]==1]["equity_per_shark"].sum()
ghazal_equity=shark_tank[shark_tank["ghazal_deal"]==1]["equity_per_shark"].sum()
```
